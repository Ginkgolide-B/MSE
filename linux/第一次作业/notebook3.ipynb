# 检查本地是否已有 SSH 密钥
!ls ~/.ssh
生成 SSH 密钥对（ed25519 安全算法）
# 请替换为【你的 GitHub 注册邮箱】，直接运行
!ssh-keygen -t ed25519 -C "your_github_email@example.com"
运行后连续按 3 次回车（默认路径、不设置私钥密码）
生成文件：
私钥：~/.ssh/id_ed25519
公钥：~/.ssh/id_ed25519.pub
启动 SSH 代理并添加私钥
# 启动 SSH 代理
!eval "$(ssh-agent -s)"

# 添加私钥到系统代理
!ssh-add ~/.ssh/id_ed25519
出现 Identity added 即成功
复制公钥内容（分系统执行，二选一）
🖥️ Mac 系统
!pbcopy < ~/.ssh/id_ed25519.pub
🖥️ Windows 系统
!clip < ~/.ssh/id_ed25519.pub
🖥️ Linux 系统
!xclip -sel clip < ~/.ssh/id_ed25519.pub
运行后公钥已自动复制到剪贴板
网页端配置公钥到 GitHub（手动操作）
打开 GitHub → 右上角头像 → Settings
左侧栏 → SSH and GPG keys → 点击 New SSH key
Title：填设备名（如 My Laptop）
Key：直接粘贴剪贴板里的公钥
点击 Add SSH key，输入 GitHub 密码确认
测试 SSH 连接是否成功
# 直接运行测试连接
!ssh -T git@github.com
首次运行提示 (yes/no)? → 输入 yes 回车
出现以下内容即为配置成功：
Hi 你的用户名! You've successfully authenticated...
免密克隆 GitHub 仓库（测试用）
# 替换为你的仓库 SSH 地址，直接克隆
!git clone git@github.com:你的用户名/仓库名.git
重要提醒
本代码仅适用于 本地 Jupyter Kernel（远程服务器 Kernel 需对应服务器操作）
私钥 id_ed25519 严禁泄露、严禁上传 GitHub
若报错 Permission denied，重新执行第 3 步添加私钥即可
